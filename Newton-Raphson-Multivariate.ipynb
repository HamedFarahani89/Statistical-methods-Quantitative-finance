```python
import numpy as np
import math
mu, sigma = 0, 5
data = np.random.normal(mu, sigma, 1000)




N = len(data)


def dsLGsian(mu,sigma):
    return -N/sigma + (sum((data-mu)**2)/sigma**3)

def ddsLGsian(mu,sigma):
    return N/sigma**2 - 3*(sum((data-mu)**2)/sigma**4)

def dmLGsian(mu,sigma):
    return sum(data-mu)/sigma**2

def ddmLGsian(mu,sigma):
    return -sum(data)/sigma**2
 
def ddmsLGsian(mu,sigma):
    return -2*sum(data-mu)/sigma**3

def Grad (mu,sigma):
    return [dmLGsian(mu,sigma),dsLGsian(mu,sigma)]

def GradMatrix (mu,sigma):
    return [ddmLGsian(mu,sigma),ddmsLGsian(mu,sigma)],[ddmsLGsian(mu,sigma),ddsLGsian(mu,sigma)]
```
