# Exponential Distribution

Anatical solution for exponential dist.


Hazard rate of it should be a constant of dist, here l.


```python
import matplotlib.pyplot as plt
import numpy as np

def PDFe(t, l):
    return  l* np.exp(-l*t)

def CDFe(t, l):
    return 1 - np.exp(-l*t)

def hazareF(t, l):
    return PDFe(t, l)/(1-CDFe(t, l))
```

```python
t = np.arange(0,5,0.1)
l = 1
plt.plot(t,PDFe(t, l))
plt.plot(t,CDFe(t, l),'r')
plt.plot(t,hazareF(t, l),'b')
hazareF(t, l)
```
