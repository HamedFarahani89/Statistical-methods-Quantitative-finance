```python
import math

def fff(x):
    return math.cos(x)-x**3

def ggg(x):
    return -math.sin(x)-3*x**2

tol = 0.0001
iter = 1000

i=0
er= 1.0
x_old= 1

while (i < iter) and (er > tol):
    x_new = x_old - fff(x_old)/ggg(x_old)
    er = abs(x_old-x_new)
    x_old = x_new
    i += 1
    print(i,x_new)
```
