### Definition:


$ds = \theta (\mu-s) dt + \sigma dW_{t}$

using discretization:

choose $s_{0}$

$s_{t+1} - s_{t} = \theta (\mu - s_{t}) dt + \sigma  dW_{t}$

$s_{t+1} = s_{t} + \theta (\mu - s_{t}) dt + \sigma dW_{t}$

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Improved O_U_process function
def O_U_process(S0, mu, theta, sigma, steps, total_time=1.0):
    dt = total_time / steps
    St = np.zeros(steps)
    St[0] = S0
    for t in range(1, steps):
        dW = np.random.normal(0, np.sqrt(dt))
        St[t] = St[t-1] + theta * (mu - St[t-1]) * dt + sigma * dW
    return St

# Set parameters for the processes
steps = 5000  # Number of steps (you can change this to see the effect)
total_time = 3.0  # Total time span

S0_values = [10, -10, 0, 0]
mu_values = [0, 0, 0, -10]
theta = 1.2
sigma = 1.4

# Generate the processes with consistent parameters
St = O_U_process(S0=S0_values[0], mu=mu_values[0], theta=theta, sigma=sigma, steps=steps, total_time=total_time)
St1 = O_U_process(S0=S0_values[1], mu=mu_values[1], theta=theta, sigma=sigma, steps=steps, total_time=total_time)
St2 = O_U_process(S0=S0_values[2], mu=mu_values[2], theta=theta, sigma=sigma, steps=steps, total_time=total_time)
St3 = O_U_process(S0=S0_values[3], mu=mu_values[3], theta=theta, sigma=sigma, steps=steps, total_time=total_time)

# Create the time vector for plotting
time = np.linspace(0, total_time, steps)

# Create the figure
fig = plt.figure(figsize=(12, 8))

# Plot the processes
plt.plot(time, St, 'b', lw=2, label='St: S0=10, mu=0, theta=1.2, sigma=1.4')
plt.plot(time, St1, 'g', lw=2, label='St1: S0=-10, mu=0, theta=1.2, sigma=1.4')
plt.plot(time, St2, '', lw=2, label='St2: S0=0, mu=0, theta=1.2, sigma=1.4')
plt.plot(time, St3, 'r', lw=2, label='St3: S0=0, mu=-10, theta=1.2, sigma=1.4')

# Plot the horizontal line for mu
plt.axhline(y=0, color='r', linestyle='-', label='mu=0')

# Set the limits for the x-axis
plt.xlim([0, total_time])

# Add the legend
plt.legend()

#
```
